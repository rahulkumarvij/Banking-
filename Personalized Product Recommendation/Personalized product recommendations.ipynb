```python
import pandas as pd
import numpy as np

# Load the dataset
data = pd.read_csv("train_ver2.csv.zip")
```

```python
data.head()
```

```python
# Select the columns of interest
last_n_column  = data.iloc[: , -24:].columns.to_list()
last_n_column = ['ncodpers'] + last_n_column
data = data[last_n_column]
```

```python
data.head()
```

```python
# Group by customers and sum the product columns to get the list of products each customer has
data = data.groupby(['ncodpers'], as_index=False).sum()
```

```python
data.head()
```

```python
# Split the data into features and labels
features = data
labels = data['ncodpers']
```

```python
# Normalize the features
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_features = scaler.fit_transform(features)
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(scaled_features, labels, test_size=0.2, random_state=0)
```

```python
# Build the recommendation model using Cosine Similarity
from sklearn.metrics.pairwise import cosine_similarity
#cosine_sim = cosine_similarity(x_train, x_train)
```

```python
x_test[0:10].shape
range(0,len(x_test[0:10]))
```

```python
for i in range(0,len(x_test[0:10])):
    print(x_test[i])
    cosine_sim_new = cosine_similarity([x_test[i]], x_train)
    indices = np.argsort(cosine_sim_new)[0][::-1][:5]
    recommended_products = np.where(x_train[indices].sum(axis=0) > 0, 1, 0)
    print(recommended_products)
    for j in range(1,len(recommended_products)):
        if recommended_products[j] == 1:
            print(features.columns[j])
    print("***********************************")
```
