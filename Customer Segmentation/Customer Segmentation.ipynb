```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the data into a Pandas dataframe
df = pd.read_csv("Churn_Modelling.csv")
```

```python
# Preprocess the data by selecting relevant features
X = df[['Age', 'Balance', 'NumOfProducts', 'HasCrCard', 'IsActiveMember', 'EstimatedSalary']]

# Handle missing values if any
X.fillna(X.mean(), inplace=True)

# Scale the features
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
# Fit a KMeans model to the data
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)

# Predict the cluster labels for each sample
labels = kmeans.predict(X)

# Add the cluster labels as a new column in the dataframe
df['Cluster'] = labels
```

```python
plt.scatter(df['Age'], df['Exited'], c=df['Cluster'])
plt.show()
```
